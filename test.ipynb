```python
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
```

```python
def preprocess_mask_labels(mask: np.ndarray):
    mask_WT = np.zeros(mask.shape)
    mask_WT[mask == 2] = 1
   
    mask_TC = np.zeros(mask.shape)
    mask_TC[mask == 1] = 1

    mask_ET = np.zeros(mask.shape)
    mask_ET[mask == 3] = 1
    
    mask_BG = np.zeros(mask.shape)
    mask_BG[mask == 0] = 1

    mask = np.stack([mask_WT, mask_TC, mask_ET, mask_BG])
    # mask = np.moveaxis(mask, (0, 1, 2), (0, 2, 1))
    return mask
```

```python
imgs_path = "BraTs/165"
```

```python
mask = np.array (Image.open (f'{imgs_path}/seg.jpg'))
```

```python
mask = (mask / 255 * 4).round ()
```

```python
mask = preprocess_mask_labels(mask)
```

```python
plt.imshow(256 - (mask[0] * 84), cmap='gray')
```

```python
plt.imshow(256 - (np.argmax(mask, axis=0) * 84), cmap='gray')
```
